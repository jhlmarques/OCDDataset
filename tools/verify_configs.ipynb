```python
import os
import pathlib
import pandas as pd
from mmengine.config import Config
from typing import NamedTuple
```

```python
PATH_WORK_DIRS = '/home/jose/Programas/OCDDataset/mmrotate-1.x/work_dirs'
EXCLUDE_DIRS = ['dota', 'pretrain', 
'rotated_rtmdet_l-6_rr_1x-ocd-cropped', 'rotated-fcos-hbox-le90_r50_fpn_psc_rr_1x_ocd-cropped',
'rotated-reppoints-qbox_r50_fpn_1x_ocd-cropped']
NUM_CONFIGS = 14
```

```python
path = pathlib.Path(PATH_WORK_DIRS)
config_files = []
for dir in path.iterdir():
    basename = os.path.basename(dir)
    if basename in EXCLUDE_DIRS:
        continue
    
    config_file = os.path.join(dir, basename + '.py')
    config_files.append(config_file)

assert len(config_files) == NUM_CONFIGS
```

```python
class ConfigData(NamedTuple):
    config_name : str
    num_batches : int
    learning_rate : float

def getConfigData(config_path):
    config_dict = Config.fromfile(config_path)
    return ConfigData(
        os.path.basename(config_path)[:-3],
        config_dict['train_dataloader']['batch_size'],
        config_dict['optim_wrapper']['optimizer']['lr'],
    )

config_data = []
for config_file in config_files:
    config_data.append(getConfigData(config_file))

df = pd.DataFrame(config_data)
df
```
