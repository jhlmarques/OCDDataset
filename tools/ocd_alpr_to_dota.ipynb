Converte anotações no formato ALPR para DOTA

```python
import os
import shutil
from PIL import Image

TRAIN_ANN_DIR = '/home/jose/Programas/OCD-Evaluation/data/OCD/train/annfiles'
TRAIN_IMG_DIR = '/home/jose/Programas/OCD-Evaluation/data/OCD/train/images'

TEST_ANN_DIR = '/home/jose/Programas/OCD-Evaluation/data/OCD/test/annfiles'
TEST_IMG_DIR = '/home/jose/Programas/OCD-Evaluation/data/OCD/test/images'
```

```python
train_alpr_dir = os.path.join(TRAIN_ANN_DIR, 'alpr')
test_alpr_dir = os.path.join(TEST_ANN_DIR, 'alpr')

train_dota_dir = os.path.join(TRAIN_ANN_DIR, 'dota2class')
test_dota_dir = os.path.join(TEST_ANN_DIR, 'dota2class')

if os.path.exists(train_dota_dir):
    shutil.rmtree(train_dota_dir)
if os.path.exists(test_dota_dir):
    shutil.rmtree(test_dota_dir)

os.mkdir(train_dota_dir)
os.mkdir(test_dota_dir)
```

```python


def convertALPRtoDOTA(img_dir, alpr_dir, dota_dir, use_two_classes):
    for ann in os.listdir(alpr_dir):
        dota_ann = os.path.join(dota_dir, ann)
        alpr_ann = os.path.join(alpr_dir, ann)

        img_path_jpg = os.path.join(img_dir, ann[:-3] + 'jpg')
        img_path_png = os.path.join(img_dir, ann[:-3] + 'png')
        if not os.path.exists(img_path_png) and os.path.exists(img_path_jpg):
            Image.open(img_path_jpg).save(img_path_png)
            
        w, h = Image.open(img_path_png).size

        with open(alpr_ann, 'r') as f_alpr, open(dota_ann, 'w') as f_dota:
            for line in f_alpr:
                cols = line.split(sep=',')[1:]
                x1x2x3x4 = cols[:4]
                y1y2y3y4 = cols[4:-1]
                
                for i, (x, y) in enumerate(zip(x1x2x3x4, y1y2y3y4)):
                    cols[i*2] = str(int(w * float(x)))
                    cols[(i*2)+1] = str(int(h * float(y)))
                
                if use_two_classes:
                    ann_class = cols[-1].strip()
                    if ann_class not in ['NormalCell', 'RoundCell']:
                        print(f'Replacing {ann_class}')
                        cols[-1] = 'NormalCell'
                    else:
                        cols[-1] = ann_class
                else:
                    cols[-1] = 'NormalCell'
                
                cols.append('0')
                line = ' '.join(cols)
                f_dota.write(line + '\n')
```

```python
print('Converting train anns...')
convertALPRtoDOTA(TRAIN_IMG_DIR, train_alpr_dir, train_dota_dir, True)
print('Converting test anns...')
convertALPRtoDOTA(TEST_IMG_DIR, test_alpr_dir, test_dota_dir, True)
```
